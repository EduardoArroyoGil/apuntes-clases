# ¿Qué es? 

Es un sistema de control de versiones. 

- Nos permite controlar nuestro trabajo.


- Pero también colaborar con otros programadores trabajando todos de forma conjunta. 

🚨🚨 Vamos a poder volver atrás en nuestro código del pasado. Pero ojo, una vez que volvemos al pasado no podremos volver hacia delante.  🚨🚨

## Ventajas de usar Git 

- **Código abierto y gratuito**:  al ser un código abierto, puedes descargar su código fuente y también realizar cambios según tus necesidades.



- **Copia de seguridad**: no nos tenemos que volver locos con miles de copias, una en un disco duro, otra en corereo, en el ordenador, etc. GitHub nos los guardará todo eternamente. 


- **Compartir código**: facilidad a la hora de compartir código, ya que simplemente subiendo tu código al repositorio principal, cualquier compañero que trabaje en el mismo proyecto podrá actualizar su copia local y tener esos archivos.


- **Deslocalización del código**: podemos estar trabajando en un mismo archivo desde distintas localizaciones y todos comparten un repositorio en común.



## Diferencias entre Git and GitHub 

- Git es una herramienta de control de versiones (software) para seguir los cambios en el código fuente.


- GitHub es un servicio en la nube basado en la web para alojar tu código fuente (repositorios Git). Es un sistema centralizado.


- Git no requiere GitHub pero GitHub requiere Git.

# ¿Cómo configuramos Git en nuestra terminal? 

Vale pero... como sé si tengo `git` instalado en mi ordenador? Ejecutaremos el siguiente comando: 

```
git version
```


Una vez que hemos confirmado que tenemos Git en nuestras terminales, vamos a vincular nuestra cuenta de Git con la de GitHub: 


- `git config --global user.name "autor"`: define el nombre del "autor" de todas las acciones que se hagan desde la terminal


- `git config --global user.email <correo>`: define el "correo" del autor de todas las acciones para el usuario definido previamente


- `git config --list`: para ver la configuración de mi Git, además confirmaremos que está bien hecha la configuración. 


- `git config --global credential.helper store`:  guardamos las credenciales creadas anteriormente para no tener que estar metiendolas cada vez que hagamos un acción en la terminal. 

# Conceptos y Comandos básicos

Antes de empezar, entendamos como esta dividido Git:

- `Workspace` o `Working directory`: es la carpeta donde estoy trabajando. 


- `Staging area`: es aquella zona donde vamos a tener aquellos ficheros a los que les hemos aplicado unas modificaciones y hemos decidido incluirlas en nuestro repositorio. 


- `Local repository`: es donde se van a guardar nuestros cambios. 

## Conceptos Básicos

- `repository| repo (repositorio|repo)`: la carpeta de trabajo.  Este repositorio registra todos los cambios realizados en los archivos de tu proyecto, construyendo el historial de versiones. 



- `remote repo (repositorio remoto)`: es un repositorio de Git que esta almacenado en la nube, por ejemplo GitHub


- `local repo (repositorio local)`: es un repositorio de Git que esta almacenado en nuestro ordenador



- `fork`: creará una copia completamente nueva en nuestro repositorio remoto (en GitHub), por lo tanto tendremos el control completo del código



- `clone (clonar)`: extrae una copia exacta de todos los datos que tenemos del repo remoto en ese momento en nuestro ordenador, incluyendo todas las versiones para cada archivo y carpeta para el proyecto.






- `untracked (no rastreado)`: significa que no esta bajo el control de versiones. Es decir, Git no esta haciendo un seguimiento de los cambios realizados en ese archivo. 


- `tracked (rastreado)`: el archivo esta bajo el control de versiones (consecuencia del comando `git add`).


- `commit`: es una forma de registrar un conjunto de cambios en el control de versiones. Puede ir desde el más pequeño cambio en un solo archivo hasta docenas que afectan a varios archivos. Un commit hace permanentes esos cambios. Por lo tanto, es un punto en la historia de un repositorio. Tiene un nombre o id que lo identifica de forma única y puede tener un mensaje que describe el significado de los cambios.



- `push (impulsar)`: envia los `commits` de nuestro trabajo en local a el repositorio remoto de GitHub. 



- `pull (sacar)`: actualiza la versión local de un repositorio desde uno remoto



- `pull request (solicitud de retirada)`: es un método para solicitar que fusione tus contribuciones a un proyecto que esta en desarrollo.


>📝 NOTA: este es el método que usaremos a lo largo del bootcamp para enviar los *labs*. 


- `branches (ramas)`: muchas veces querremos trabajar en nuestro proyecto sin tener que trastear con el trabajo que has hecho hasta ese momento. ¡En ese caso, una buena cosa es crear una rama! 

    Una rama es una desviación del curso de la historia reflejada por la rama principal (*master* o *main*). La rama principal y la nueva pueden evolucionar de forma diferente desde ese punto de la historia. En algún momento en el futuro podrías fusionarlas de nuevo, pero, por ahora, puedes cambiar cosas en una de las ramas sin afectar a la otra.





- `merge (mergear)`: nos permite unir los cambios hechos en una rama en integrarlas en un única rama (normalmente la principal) 


- `.gitignore`: es un archivo de texto que indica a Git qué archivos o carpetas debe ignorar en un proyecto. Normalmente se utiliza para evitar subir al repo remoto los archivos temporales de nuestro repositorio local de trabajo que no son útiles para otros colaboradores, como los archivos temporales que crean los IDE, etc.



## Comandos Básicos

- `git init <carpeta>`:  crea un repositorio vacío en "carpeta". Sin argumentos permite inicializar la carpeta actual como repositorio. 


- `git status`: muestra la lista de archivos de los cambios realizados sin *trackear* y los *trakeados* 


- `git clone <ulr del repositorio>`: clona el "repositorio" en nuestro ordenador, es decir, bajarse una copia completa del mismo a nuestro a ordenador. 


- `git add <carpeta|archivo>`: agrega los cambios realizados en dicha carpeta o archivo para la siguiente confirmación. 



- `git rm --cached <nombre del fichero>`: me elimina de los archivos *trackeados*. 


- `git commit -m <mensaje>`: confirma los cambios en esa carpeta, usa el "mensaje" como una etiqueta que nos ayudará si queremos volver a una vesión anterior. 🚨 Hay que poner mensajes que nos indiquen que es lo significativo del cambio. 


- `git push origin <rama>`: envía a la "rama" al repositorio actualizando los cambios que tenemos en local en la "nube" de GitHub. 


- `git pull <remoto>`: obtiene una copia del repositorio "remoto" y lo combina con el repositorio local. 


- `git log`: permite ver todos los *commits* hechos con su código identificador que será el que nos ayude a realizar cualquier cambio dentro de git. 


- `git reset --hard <commit-id>`: volvemos a un commit anterior

# Trabajando con Git 

Ahora es el momento de crearnos nuestro propio repositorio y poner en práctica todos los conceptos que hemos ido viendo hasta ahora 🙇🏼‍♀️🙇🏽‍♂️

## Creamos un repo

Podemos crear un repo de dos formas diferentes:

- De forma local y vincularlo con GitHub, usando la terminal con los conceptos que aprendimos el martes 


- De forma remota y descargandonoslo en nuestro ordenador


### De forma local 

```           
mkdir aprendiendo_git # creamos una carpeta nueva

cd aprendiendo_git # accedemos a la carpeta 

git init # pero esto no es lo más recomendable

ls -a # comprobamos que es un repo. TENDREMOS QUE VER UN ARCHIVO .GIT
```




```
git remote add origin <url_del_repo> # vinculamos el repo en local con el repo remoto
```



### De forma remota

```
# creamos en repo nuevo en GitHub

git clone <url del repo> # clonamos el repo, es decir, nos hacemos una copia exacta de un repo que esta en GitHub
```

## Agregamos archivos trabajando en local

Vamos a crear unos ficheros nuevos sobre los que iremos trabajando a lo largo de este ejemplo:

```
touch pruebas.ipynb # creamos un archivo nuevo
```



```
git status #chequeamos los cambios en nuestro repo.
```

> El estado inicial de los archivos nuevos es *untracked* y lo veremos en **color rojo**



El siguiente paso es "decirle" a git que ficheros queremos que considere, es decir, pasaremos a la `staging area` 
```
git add pruebas.ipynb

📌 También podremos ejecutar el comando  `git add .` para agregar todos los archivos creados y/o modificados
```

> Ejecutando `git status` veremos que los ficheros añadidos apararecen en **color verde**, ahora estos ficheros estarán en el *staging area*



Por último, tenemos que guardar los cambios definitivamente en nuestro repo local. 

``` 
# 🚨 Recordamos que este mensaje debe ser lo más explicativo que podamos para que en caso de que tengamos que volver a un commit anterior sepamos donde queremos volver.

git commit -m "creando el primer cuaderno de trabajo"
```

> Si ejecutamos ahora `git status` ahora ya no nos aparece el archivo pruebas. 



⚠️ Hasta aquí tenemos nuestros datos en nuestro repositorio local

## Subimos nuestros ficheros de local a remoto

```
git push origin main # para subir nuestros ficheros de local a remoto (GitHub) 
```




📌 GENIAL! Ya tenemo nuestro repo nuevo 🤓. Si ahora nos vamos al repo de GitHub veremos los cambios en nuestro repo, con el nombre del commit que hemos puesto en los pasos anteriores

## Hagamos un ejemplo de un pull

Actualizar un repo forkeado que no es mío (por ejemplo, los apuntes)

🚨🚨 Este será nuestro mantra cada día para actualizar los apuntes de clase 🚨🚨

Para hacer esto tenemos que estar en la carpeta de apuntes de clase. Una vez en la carpeta ejecutamos el siguiente comando:

```
git remote add upstream https://github.com/Ironhack-Part-Time-Enero2022/apuntes-clases.git 
```

   🚨⚠️ **Este paso solo lo haremos una vez** 🚨⚠️




```
git pull upstream main  # Bajo a local los cambios que se han hecho en la carpeta de apuntes de clase para ver el jupyter que toca hoy
```

🚨📌 FALTA UNA CAPTURA, LO HAGO MAÑANA CON LA ORGANIZACION DEL BOOTCAMP DE FULL

Ahora ya tenemos los cambios en nuestro ordenador. Estos cambios están en nuestro ordenador, el siguiente paso, subirlo a GitHub (nuestro repositorio remoto) como hemos aprendido. 


```
git add .  # Suele estar prohibido pero en este caso aceptaremos pulpo como animal de compañia

git commit -m "actualizo repo"

git push origin main --> Subo a Github lo que me he traído al ordenador

```

# Trabajando con ramas 

Son lineas de tiempo de trabajo. Es decir, una rama  es como la creación de una nueva isla basado en la situación actual del repo original, esta isla respeta las mismas leyes, pero será completamente distinta. 


**¿Para qué podemos usar las ramas?**

> Para crear código pero que no queremos que la rama principal se vea afectada.   
>Imaginemos, por ejemplo, que queremos trabajar en algo que no es definitivo y no lo queremos subir a master por si rompemos algo que funciona. En este caso crearíamos una rama aparte en la que trabajar, y cuando nuestro trabajo esté listo, la fusionaríamos con master.

## Comandos importantes

Antes de empezar veamos cuales son los principales comandos que usaremos cuando trabajamos con ramas:

- `git branch trabajo-adicional`: creamos una rama nueva con el nombre "trabajo_adicional". 


- `git branch`: nos muestra en que rama estamos. 


- `git checkout trabajo-adicional`: nos movemos a la nueva rama


Podemos unir la rama master o main con la nueva rama creada, para eso **tenemos que estar en la rama master o main** y ejecutar el siguiente comando: 

- `git merge trabajo-adicional`: combinamos el código de la rama creada con la main o master. 


Si queremos eliminar la rama creada después de mergear: 

- `git branch -d trabajo_adicional`: eliminamos una rama


> ⚠️ Atajo ⚠️ Si queremos crear y movernos a la rama creada  
> `git checkout -b trabajo_adicional`

## Manos a la obra!

```
# lo primero de todo, chequeamos en que rama estamos: 
git branch # como no tenemos ninguna rama, nos tenmdrá que salir que estamos en la main. 

# ahora vamos a crearnos una rama nueva
git branch aprendiendo

# nos movemos a la rama nueva
git checkout aprendiendo

#creamos un nuevo fichero 
touch nuevo.ipynb
```
